```python
#Importamos librerias
import pandas as pd
import seaborn as sns

import matplotlib.pylab as plt
import numpy as np
```

```python
#Leemos el archivo
newcars = pd.read_csv("American New Cars of 1993.csv",na_values = ":")
newcars
```

```python
#Se crea el plot scatter
x = newcars.Cylinders
y = newcars.Horsepower
plt.scatter(x,y, color="blue")
plt.xlabel("Cylinders")
plt.ylabel("Horsepower")
```

```python
#Se crea una grafica de barras con la variable de Cylinders
sns.set(rc={"figure.figsize":(11.7,8.27)})
sns.displot(newcars["Cylinders"], bins=30)
plt.show()
```

```python
#Se crea un mapa de calor
correlation_matrix = newcars.corr().round(2)
sns.heatmap(data=correlation_matrix, annot=True)
```
